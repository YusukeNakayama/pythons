```python
import psycopg2

##env
host=%env PG_HOST
port=%env PG_PORT
db=%env PG_DB
user=%env PG_USER
password=%env PG_PASS

pgconfig = {
    'host':host,
    'port':port,
    'database':db,
    'user':user,
    'password':password,
}

con = psycopg2.connect(**pgconfig)
cur = con.cursor()
cur.close()
```

```python
%load_ext sql

##env
host=%env PG_HOST
port=%env PG_PORT
db=%env PG_DB
user=%env PG_USER
password=%env PG_PASS

pgconfig = {
    'host':host,
    'port':port,
    'database':db,
    'user':user,
    'password':password,
}
dsl = 'postgres://{user}:{password}@{host}:{port}/{database}'.format(**pgconfig)
%sql $dsl
```
